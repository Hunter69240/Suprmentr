```python
'''
y machine need numbers?
we have to convert text ot numbers

process of converting text to numbers is called text vectorization
so ML models can learn patterns

ex: I Love Ai
afte vectorization -> [1, 0, 1] (1 for love and ai, 0 for i)

2 popular techniques:
1 Bag of Words (BoW)
TF-IDF
'''
```

```python
'''
Bag of words
ex:Count how many times word appears

I love nlp => doc 1
i love machine learning => doc 2
nlp is fun => doc 3

step 1 : build vocabulary - collect all unique words
[i, love, nlp, machine, learning, is, fun]

step 2 count word frequency

docs  i love nlp
docs  1  1    1
docs1 1  1    0
docs2 1  1    0
docs3 0  0    1

this is called document term matrix

key idea:
BOW ignores - grammer and wrong words

ex - i love nlp and love nlp i (both become same vector)
'''
```

```python
from sklearn.feature_extraction.text import CountVectorizer

documents=[
    "I love NLP",
    "I love machine learning",
    "NLP is fun"
]

vectorizer = CountVectorizer()
x=vectorizer.fit_transform(documents)
print("Vocabulary:")
print(vectorizer.get_feature_names_out())

print("\nVector representation:")
print(x.toarray())
```

```python
'''
TF-IDF 

TF-term frequence
ex: AI AI ML
word    count
AI      2
ML      1

IDF - Inverse document frequence
Measure word rarity across documents
IDF=log(total number of documents / number of documents containing the word)
if word appears in many documents, it gets lower weight
rare words get higher weight
'''
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
documents=[
    "I love NLP",
    "I love machine learning",
    "NLP is fun"
]

vectorizer = TfidfVectorizer()
x=vectorizer.fit_transform(documents)
print("Vocabulary:")
print(vectorizer.get_feature_names_out())

print("\nTF-IDF representation:")
print(x.toarray())
```
