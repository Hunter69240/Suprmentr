```python
import random
from collections import Counter
text='i love coding i love ai i love chocolate ai is futute coding is fun'
words=text.split()
model={}
for i in range(len(words)-1):
    word=words[i]
    next_word=words[i+1]

    if word not in model:
        model[word]=[]
    model[word].append(next_word)

def predict_next(word):
    if word in model:
        return random.choice(model[word])
    else:
        return 'unknown'
    
#generate sentence
def generate_text(start_word,length=5):
    result=[start_word]

    for _ in range(length):
        next_word=predict_next(result[-1])
        result.append(next_word)

        return ' '.join(result)
def predict_smart(word):
    if word in model:
        next_words=model[word]
        most_common=counter(next_words).most_common(1)
        return most_common[0][0]
    else:
        return 'unknown'
#test
print(generate_text('ai'))


```
