```python
'''
What is decision tree?
A decision tree is supervised ml algorithm , used for classification or regression

it splits data into branches based on feature conditions.

How it works:
Step 1:Choose best features to split.
using:gini index,entropy(information gain)

step 2: Split dataset

step 3:Repeat recursivelu

Step 4 : stop when:
*pure node
* max depth reached
*minimum samples reached

entropy formula ( for classification)

measure impurity in dataset

entropy=plog2(p)-(1-p)log2(1-p)

if all same class ->entropy=0
if 50-50 ->entropy=1(max impurity)


'''
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
iris=load_iris()
```

```python
X=iris.data
y=iris.target
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
#create model
model=DecisionTreeClassifier(criterion="gini",max_depth=3)
```

```python
#train
model.fit(X_train,y_train)
```

```python
#predict
y_prediction=model.predict(X_test)
```

```python
#accuracy
print("Accuracy:",accuracy_score(y_test,y_prediction))
```

```python
plt.figure(figsize=(12,8))
tree.plot_tree(model,feature_names=iris.feature_names,class_names=iris.target_names,filled=True)
plt.show()
```

```python
'''
underfitting:
model is too simple
ex:depth=1
result:low train accuracy,low test accuracy
graph:straight line trying to fit curved line


overfitting:
model is too complex
ex=max_depth=None
result=100% training accuracy,poor test accuracy
why:it memorizes training data
'''
```

```python
'''
Random Forest:
problem in DT: high varaiance , overfits easily
solution?build many trees -> Random forest

what is Random Forest?
Random Forest is collection of decision tree
Uses:
Bootstrap sampling, random feature selection,majority voting

working flow:
step 1: create 100 random samples from datasets
step 2:train 100 DT
step 3:each tree predicts
step 4:Final prediction=majority voting
'''
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
#create model
rf_model=RandomForestClassifier(
    n_estimators=100,
    max_depth=5,
    random_state=42
)
#train
rf_model.fit(X_train,y_train)

#predict
y_pred_rf=rf_model.predict(X_test)

#accuracy
print("Random forest accuracy:",accuracy_score(y_test,y_pred_rf))
```
