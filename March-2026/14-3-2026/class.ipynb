# Lemminazations and Tokenization

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
'''
Tokenization 
Breaking text into smaller pieces called as tokens

ex: I Love NLP
token : ['I', 'Love', 'NLP']

why:
ML algo process individual units of text.

types of tokenization:
1.Word Tokenization:Break sentence into words
2.Sentence Tokenization:Break text into sentences
'''
```

```python
from nltk.tokenize import word_tokenize
text="I Love watching movies on weekends"
tokens=word_tokenize(text)
print(tokens)
```

```python
text1="I Love watching movies on weekends. I love NLP"
tokens1=word_tokenize(text1)
print(tokens1)
```

```python
'''
Problrm with tokenization: It produces many unwanted words
so we remove this using stop words
'''
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

text="The movie was very good and interesting"
tokens=word_tokenize(text)
stop_words=set(stopwords.words('english'))
filtered_words=[]

for word in tokens:
    if word.lower() not in stop_words:
        filtered_words.append(word)
print(filtered_words)
```

```python
'''
LEMMATIZATION

Converts word to its base or root form'
'''
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

lemm=WordNetLemmatizer()
text="The children are playing with toys"
tokens=word_tokenize(text)
lemmas=[]
for words in tokens:
    lemma=lemm.lemmatize(words)
    lemmas.append(lemma)
print(lemmas)
```
