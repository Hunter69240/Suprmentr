```python
'''
Word embedding ?
A word embedding is  dense vector representation of word  where similar words have similar value

why embeddings powerful?
help models understand:
i)similarity
ii)relationship
iii)hidden patterns of text


important properties of embeddings:
1. Dense Vector : less 0's and compact representation
2. Semantic meaning : similar words have similar vectors
3. Learned automatically : model learns vector values from text corpus
'''
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
```

```python
sentences=[
    " I like dogs",
    "I Like cats",
    "Dogs and cats are pets",
    "I love pets"
]
```

```python
vectorizer = CountVectorizer()
X=vectorizer.fit_transform(sentences)
print("Vocabulary:",vectorizer.get_feature_names_out())
similarity=cosine_similarity(X)
df=pd.DataFrame(similarity,columns=[f"S{i+1}" for i in range(len(sentences))],index=[f"S{i+1}" for i in range(len(sentences))])
print("\n Sentences similarity matrix: ")
print(df)
```

```python
'''
Word2Vec:
Technique that learns word embeddings by predicting relationship between words 

Introduced by google
Learns word meaning from nearby words

Architecture:
1. CBOW (Continuous Bag of Words) : Precict target word by surrounding context words

2. Skip-Gram : 

Flow or Word2Vec:
step 1: take large text corpus
step 2: slide a context window
step 3:train a small neura network
step 4:update vector so prediction improves
step 5:final hidden layer weights become word embeddings
'''
```

```python
from gensim.models import Word2Vec
```

```python
sentences = [
    ["i", "like", "dogs"],
    ["i", "like", "cats"],
    ["dogs", "and", "cats", "are", "pets"],
    ["i", "love", "pets"],
    ["cats", "drink", "milk"],
    ["dogs", "chase", "balls"]
]
```

```python
model=Word2Vec(sentences,vector_size=50,window=2,min_count=1,sg=1)
print("Vector for dogs: ")
print(model.wv['dogs'])

print("\n Words similar to cats: ")
print(model.wv.most_similar('cats'))

'''
vector_size=50 each word represented y 50 numbers
window=2 models look at 2 words before and after target word
min_count=1 even rare words can be included

'''
```

```python
'''
GloVe (Global Vectors for Word Representation):

unlike Word2Vec which learns from local context prediction. 
GloVe uses global co-occurrence stats
that means it studies how frequently words appear together in the entire corpus

Word2Ve vs GloVe:
word2vec : local context prediction
GloVe : global co-occurrence stats, counts/stats based
'''
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

corpus=[
    'ice is cold',
    'snow is cold',
    'steam is hot',
    'fire is hot'
]

vectorizer=CountVectorizer()
X=vectorizer.fit_transform(corpus)
df=pd.DataFrame(X.toarray(),columns=vectorizer.get_feature_names_out())
print(df)
print('\n This matrix is not full Glove, but helps understand word occurance pattern')
```
