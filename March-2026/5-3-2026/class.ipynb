```python
'''
Unsupervised learning and kmeans cluster
'''

'''
unsupervised learning?
no labeled output is given,algorithm finds  the hidden pattern

types:
i) Clustering:Group similar data points

ii) Association : Find relations
'''
```

```python
'''
K-means clustering:
Groups data into k clusters
where k=number of groups

How it works:
Step 1 : Choose k
Step 2:Initialise centroid
Step 3: Assign points to nearest centroid
Step 4 : Update centroid : Average of all points in cluster
Step 5 : Repeat until stable
'''
```

```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data={
    "Age":[22,23,24,45,46,47,30,31,32],
    "Income":[25,28,26,90,88,95,60,65,62],

}

df=pd.DataFrame(data)
print(df)
```

```python
plt.scatter(df["Age"],df["Income"])
plt.xlabel("Age")
plt.ylabel("Income")
plt.title("Customer Data")
plt.show()
```

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit(df)
```

```python
#get cluster labels
labels=kmeans.labels_
print(labels)
```

```python
'''
Cluster 0 -> Young
Cluster 1 -> High income
Cluster 2 -> Middle
'''
```

```python
#plot clusters
plt.scatter(df["Age"],df["Income"],c=labels)
plt.xlabel("Age")
plt.ylabel("Income")
plt.title("Kmeans clustering Customer Data")
plt.show()
```

```python
#show centroids
centroids=kmeans.cluster_centers_
print(centroids)
```

```python
plt.scatter(df["Age"],df["Income"],c=labels)
plt.scatter(centroids[:,0],centroids[:,1],marker="X",s=200)
plt.show()
```
