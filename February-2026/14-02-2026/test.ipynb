```python
import pandas as pd
```

```python
mark=pd.Series([85,90,78,92])
print(mark)
```

```python
mark=pd.Series([85,90,78,92],index=['a','b','c','d'])
print(mark)
```

```python
print(mark['c'])
print(mark[1])
```

```python
#creating dataframe from dictionary

data={
    "name":['Ravi','Anita','John','Meera'],
    'age':[21,20,22,23],
    'marks':[85,90,78,92]
}

df=pd.DataFrame(data)
print(df)
```

```python
print(df['name'])
```

```python
print(df[['name','marks']])
```

```python
print(df.loc[0]) #by label
print(df.iloc[1]) #by index
```

```python
#adding column
df['Result']=df['marks']>80
print(df)
```

```python
#deleting
df.drop('age',axis=1,inplace=True)
print(df)
```
