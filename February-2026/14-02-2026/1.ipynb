# 📘 Pandas Student Data Analysis – Explanation Notes

---

# 1️⃣ Reading the CSV File

```python
import pandas as pd
df = pd.read_csv('students_data.csv')
print(df)
```

## Explanation
- `pd.read_csv()` loads the CSV file into a Pandas DataFrame.
- The DataFrame is a 2D table (rows and columns).
- `print(df)` displays the dataset.

---

# 2️⃣ Adding a New Column – Total Marks

```python
df['total'] = df['maths'] + df['science'] + df['english']
```

## Explanation
- A new column `total` is created.
- It performs **row-wise addition** of three subject columns.
- For each student:

```
total = maths + science + english
```

This updates the DataFrame permanently.

---

# 3️⃣ Adding Average Column

```python
df['average'] = df['total'] / 3
```

## Explanation
- Creates a new column `average`.
- Divides total marks by 3 subjects.
- Formula:

```
average = total / number_of_subjects
```

---

# 4️⃣ Top 3 Students Based on Total

```python
df.nlargest(3, 'total')
```

## Explanation
- `nlargest(n, column)` returns top `n` rows sorted by a column.
- Here it selects top 3 students with highest total marks.
- Efficient and cleaner than sorting + head().

Equivalent to:

```python
df.sort_values(by='total', ascending=False).head(3)
```

---

# 5️⃣ Department-wise Average Total Marks

```python
df.groupby('dept')['total'].mean()
```

## Explanation
- `groupby('dept')` → Groups students by department.
- `['total']` → Selects total marks column.
- `.mean()` → Calculates average total marks per department.

Formula internally:

```
Department Average = Sum of totals in dept / Number of students in dept
```

Returns a Pandas Series indexed by department.

---

# 6️⃣ Students Scoring Above 75 in All Subjects

```python
filtered_rows = df[
    (df['maths'] > 75) & 
    (df['science'] > 75) & 
    (df['english'] > 75)
]
print(filtered_rows)
```

## Explanation
- Each condition creates a True/False column.
- `&` is the AND operator for Pandas.
- Only rows where all conditions are True are selected.
- Parentheses are required for correct evaluation.

Logical Meaning:

```
maths > 75 AND
science > 75 AND
english > 75
```

---

# 7️⃣ Checking Missing Values

```python
df.isnull()
```

## Explanation
- Returns a DataFrame of True/False.
- True → value is missing (NaN).
- False → value exists.
- Used to identify missing data locations.

To count missing values:

```python
df.isnull().sum()
```

---

# 8️⃣ Handling Missing Values (Mean Imputation)

```python
df['maths'] = df['maths'].fillna(df['maths'].mean())
df['english'] = df['english'].fillna(df['english'].mean())
print(df)
```

## Explanation
- `fillna()` replaces missing values.
- `df['maths'].mean()` calculates average ignoring NaN.
- Missing values are replaced with mean of the column.

Why Mean?
- Maintains dataset size
- Preserves approximate distribution
- Common preprocessing method in ML

Mathematically:

```
New Value = (Sum of non-missing values) / (Count of non-missing values)
```
